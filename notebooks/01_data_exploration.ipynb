# 01 — Data Exploration

Load and explore market data for the ML Alpha Lab universe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
%matplotlib inline
```

```python
from src.data.data_loader import DataLoader
from src.data.data_cleaner import DataCleaner
from src.data.data_validator import DataValidator

loader = DataLoader()
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'NVDA', 'JPM', 'V', 'JNJ', 'PG']
prices = loader.load_yahoo_finance(tickers, start_date='2018-01-01', end_date='2024-01-01')
print(f'Shape: {prices.shape}')
prices.head()
```

```python
cleaner = DataCleaner()
prices_clean = cleaner.clean(prices)
returns = cleaner.compute_returns(prices_clean, method='log')
print(f'Returns shape: {returns.shape}')
returns.describe()
```

```python
validator = DataValidator()
print(validator.generate_report(prices_clean))
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8))
# Normalized prices
(prices_clean / prices_clean.iloc[0] * 100).plot(ax=axes[0], linewidth=1)
axes[0].set_title('Normalized Price Series (Base=100)')
axes[0].set_ylabel('Price Index')
# Returns distribution
returns.stack().hist(bins=100, ax=axes[1], edgecolor='none')
axes[1].set_title('Returns Distribution')
axes[1].set_xlabel('Daily Log Return')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
corr = returns.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='RdYlGn', center=0, ax=ax, square=True)
ax.set_title('Return Correlation Heatmap')
plt.tight_layout()
plt.show()
```
